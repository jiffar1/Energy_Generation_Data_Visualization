```python
ax = power_data['country'].value_counts().sort_values(ascending=False)[:20].plot.bar(x = 'Country', y = '', title = 'Electricity Generating Facilities by Country', rot = 90, figsize = (12, 10), fontsize =15)
ax.set_xlabel("Country", fontsize=15)
ax.set_ylabel("Number of Power Generating Facilities", fontsize=15)
plt.show()
```

```python
capacity = power_data.groupby('country')['capacity_mw'].sum().round(2)
```

Next, I average the capacity by plant for the world:


```python
#Next, I average the plant capacity for the world:
power_data['capacity_mw'].mean()

#Then I average it by country:
average_capacity_by_country= power_data.groupby('country_long')['capacity_mw'].mean()
average_capacity_by_country = average_capacity_by_country.sort_values(ascending=False)
print(average_capacity_by_country.head(20))

#From above code we see that the USA and China are not top on the list of having the largest average plant size:
# Paraguay is top on list followed by Kuwait and Ukraine.
```

Next I compare the average generating capacity of a Chinese plant and compare it to a USA plant.

```python
#I will now see the average generating capacity of a Chinese plant:
#And compare it to the average generating capacity of a US plant:
print('Average American plant:' + str(average_capacity_by_country['United States of America']) + 'MW')
print('Average Chinese plant: ' + str(average_capacity_by_country['China']) + 'MW')

#Like the preliminary data exploration suggested, a Chinese plant is on average about 3 times bigger than a US
#plant. That is why China generates more electricity than the US even though China has 1/3 the number of plants
#on this dataset than does the US
```

Next, I check the world electricity generation by the sources of fuel(energy).
I see from the calculation below that sadly coal is still the number one source of electricity around the world.
Followed by Natural Gas, then Hydroelectric.

```python
#Next, I check the world electricity generation by the sources of fuel(energy).
#I see from the calculation below that sadly Coal is still the number one source of electricity around the world.
#Followed by Natural Gas, then Hydroelectric.
energy_by_source = power_data.groupby('fuel1')['capacity_mw'].sum().round(2)
energy_by_source.sort_values()
```

```python
#The above data in graphical form is below
#Again from the graph we see coal dominating electricity generation across the world.
ax = capacity_by_fuel_type_graph = energy_by_source.sort_values(ascending = False).plot.bar(rot=90, title = 'World Electricity Generation by Fuel Type, 2017', figsize = (12,10), fontsize =15)
ax.set_xlabel("Fuel Type", fontsize=15)
ax.set_ylabel("Generated MW", fontsize=15)
plt.show()
```

Next, I will look at the same generation by fuel type data for China and the US.
First I look at the USA

```python
#For the USA
energy_by_source_usa = power_data[power_data['country'] == 'USA'].groupby('fuel1')['capacity_mw'].sum().round(2)
ax = energy_by_source_usa.sort_values(ascending=False).plot.bar(rot=90, title = 'USA electricity generation by fuel type, 2017', figsize = (12,10), fontsize =15)
ax.set_xlabel("Fuel Type", fontsize=15)
ax.set_ylabel("Generated MW", fontsize=15)
plt.show()
#We see that coal is still the top fuel source for the USA.
```

Now I look at China:

```python
#Now I look at China:
energy_by_source_china = power_data[power_data['country'] == 'CHN'].groupby('fuel1')['capacity_mw'].sum().round(2)
ax =energy_by_source_china.sort_values(ascending=False).plot.bar(rot=90, title = 'China\'s electricity generation by fuel type, 2017', figsize = (12, 10), fontsize =15)
ax.set_xlabel("Fuel Type", fontsize=15)
ax.set_ylabel("Generated MW", fontsize=15)
plt.show()

#We see that coal is the top fuel source for China.


#Infact, we see that coal is a much bigger and widely used fuel source in China than it is in the USA.
```

We see that coal is the top fuel source for China.

Infact, we see that coal is a much bigger share of the electric mix in China than in the USA.
